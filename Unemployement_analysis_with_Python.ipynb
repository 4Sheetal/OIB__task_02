```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px # data Visualisation
```

```python
# read the data
data=pd.read_csv('/content/Unemployment in India.csv')
data.head
data.info # get the info of data
```

```python
data.describe()
```

```python
# See if this dataset contains missing value or not
print(data.isnull().sum())
```

```python
# Drop the null values of the rows
df=data.dropna()
print(df.isnull().sum())
```

```python
# estmaited the correlation between the feature of this dataset
plt.style.use("seaborn-whitegrid")
plt.figure(figsize=(12,10))
sns.heatmap(df.corr())
plt.show()
```

```python
#Look at the estimated number of employees according to different Area of India"
df.columns=["States","Date", "Frequency","Estimated Unemployment Rate (%)","Estimated Employed","Estimated Labour Participation Rate (%)","Area"]
plt.title("Indian Unemployed rate")
sns.histplot(x="Estimated Employed",hue="Area",data=df)
```

```python
# Create a Dashboard to analyse the unempolyement rate of each indian state by region, .use sunburst plot.
# Create a Dashboard to analyze the unemployment rate of each Indian state by region, use sunburst plot.
Unemployment = df[["States", "Area","Estimated Unemployment Rate (%)"]]
figure = px.sunburst(Unemployment, path=["States", "Area"], values="Estimated Unemployment Rate (%)", width=700, height=700, color_continuous_scale="RdYlGn", title="Unemployment Rate in Indian States by Region")
figure.show()
```
